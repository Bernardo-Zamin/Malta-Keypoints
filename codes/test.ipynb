```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import wget
import re
import pandas as pd
import tqdm
```

```python
main_url = 'https://libras.cin.ufpe.br/?page='
labels = []
label = []
links = []
for i in tqdm.tqdm(range(1, 71)):
    response = requests.get(f'{main_url}{i}')
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        links += re.findall(r'<a href="(.*?)"', str(soup.body.find_all('td')))  


        labels +=  [element.text for element in soup.body.find_all('td')]
        label = [re.sub(r'\s+', ' ', name.strip()) for name in labels[::2]]
    else:
        print(f'Error {response.status_code}')
        break
```

```python
names_links = {}  
for i in range(len(label)):
    names_links[label[i]] = links[i]

len(names_links)
```

```python
df = pd.DataFrame(names_links.items(), columns=['Name', 'Link'])
df
```

```python
download_directory = '/mnt/B-SSD/bernardo/libras-ufpe'

# Certifique-se de que o diretório de download existe
os.makedirs(download_directory, exist_ok=True)

# Função para baixar os vídeos de um URL
def download_videos(term, url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    videos = soup.find_all('video')

    for i, video in enumerate(videos, start=1):
        source = video.find('source')
        if source and source['type'] == 'video/mp4':
            video_url = source['src']
            video_response = requests.get(video_url)

            # Renomear o vídeo com o termo e um número sequencial
            filename = f"{term.replace(' ', '_')}_{i}.mp4"
            file_path = os.path.join(download_directory, filename)

            # Baixar e salvar o vídeo
            with open(file_path, 'wb') as f:
                f.write(video_response.content)
            print(f"Baixado: {filename}")

# Iterar sobre cada termo e URL e chamar a função de download
for term, url in names_links.items():
    download_videos(term, url)

print("Todos os vídeos foram baixados.")
```
