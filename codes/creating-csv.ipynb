```python
import pandas as pd
import os
import locale
import re
```

```python
def sorted_path(path):

    locale.setlocale(locale.LC_ALL, '')

    file_list = os.listdir(path)

    sorted_file_list = sorted(file_list, key=locale.strxfrm)

    full_paths = [os.path.join(path, file_name) for file_name in sorted_file_list]

    return full_paths

def labels(video_paths):
    video_paths = sorted_path(video_paths)
    file_names = [os.path.basename(video_path) for video_path in video_paths]
    file_names = [file_name.rsplit('_', 1)[-1] for file_name in file_names]

    names = []
    for file in file_names:
        if file not in names:
            names.append(file)
    return names

def video_names(paths):
    video_dict = {}
    for path in paths:
        name = os.path.basename(path)[:-6]
        if name not in video_dict:
            video_dict[name] = []
        video_dict[name].append(path)
    return video_dict
```

```python
path = '/B/zamin/dataset-ufpe'
paths = sorted_path(path)
names = labels(path)

print("Verificação: ",len(paths))
print("Verificação: ",len(names))
```

```python
base_path = '/B/zamin/dataset-ufpe'
paths = sorted_path(base_path)
video_dict = video_names(paths)

data = {
    'name': [],
    'checked' : None,
    'lemma' : None,
    'path': [],
    'dictionary': 'LIBRASIL',
    'group': None,
    'actor': 1

}

for name, videos in video_dict.items():
    data['name'].append(name)
    data['path'].append(videos)

df = pd.DataFrame(data)
df.to_csv('dataset-ufpe.csv', index=False)
```
