```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Principal.csv", low_memory=False)
data = data['paths']
data.head()
```

```python
vids = []
for i in data:
    if i.startswith('['):
        i = i[1:-1]
    if i.endswith(']'):
        i = i[:-1]
    vids += list(i.split(','))


for i in vids:
    print(i)
    
```

```python
df = pd.DataFrame(vids, columns=['Paths'])
df.to_csv('Paths_only.csv', index=False)
```
